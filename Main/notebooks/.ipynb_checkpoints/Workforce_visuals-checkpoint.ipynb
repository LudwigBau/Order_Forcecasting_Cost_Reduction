## Custom functions 

```python
def calc_savings(df, group=False, scenario=None, week=False):
    
    if group == True and week == False:
        savings_df_list = []  # Initialize an empty list to hold DataFrames

        # calculate rel cost, backlog and overcap per scenario (e.g. there are 3 cost scenarios)
        for i in df[scenario].unique():
            
            group_df = df[df[scenario] == i].copy()

            actual_cost = df[(df.model == 'actual') & (df[scenario] == i)]['avg_cost'].values[0]
            bench_cost = df[(df.model == 'L_4_sarimax') & (df[scenario] == i)]['avg_cost'].values[0]
            actual_backlog = df[(df.model == 'actual') & (df[scenario] == i)]['avg_backlog'].values[0]
            bench_backlog = df[(df.model == 'L_4_sarimax') & (df[scenario] == i)]['avg_backlog'].values[0]
            bench_overcap = df[(df.model == 'L_4_sarimax') & (df[scenario] == i)]['avg_overcap'].values[0]

            group_df["rel_actual_cost"] = (group_df.avg_cost / actual_cost) - 1
            group_df["rel_bench_cost"] = (group_df.avg_cost / bench_cost) - 1
            group_df["rel_actual_backlog"] = (group_df.avg_backlog / actual_backlog) - 1
            group_df["rel_bench_backlog"] = (group_df.avg_backlog / bench_backlog) - 1
            group_df["rel_bench_overcap"] = (group_df.avg_overcap / bench_overcap) - 1

            savings_df_list.append(group_df)

        savings_df = pd.concat(savings_df_list, ignore_index=True)
    
    # sort for week and scenario
    elif group == True and week == True:
        
        savings_df_list = []
        
        # loop over weeks and scenario
        for week_no in df["week"].unique():
            for i in df[scenario].unique():
            
                group_df = df[(df[scenario] == i) & (df["week"] == week_no)].copy()
        
                actual_cost = df[(df.model == 'actual') & (df[scenario] == i) & (df["week"] == week_no)]['avg_cost'].values[0]
                bench_cost = df[(df.model == 'L_4_sarimax') & (df[scenario] == i) & (df["week"] == week_no)]['avg_cost'].values[0]
                actual_backlog = df[(df.model == 'actual') & (df[scenario] == i) & (df["week"] == week_no)]['avg_backlog'].values[0]
                bench_backlog = df[(df.model == 'L_4_sarimax') & (df[scenario] == i) & (df["week"] == week_no)]['avg_backlog'].values[0]
                bench_overcap = df[(df.model == 'L_4_sarimax') & (df[scenario] == i) & (df["week"] == week_no)]['avg_overcap'].values[0]
    
                group_df["rel_actual_cost"] = (group_df.avg_cost / actual_cost) - 1
                group_df["rel_bench_cost"] = (group_df.avg_cost / bench_cost) - 1
                group_df["rel_actual_backlog"] = (group_df.avg_backlog / actual_backlog) - 1
                group_df["rel_bench_backlog"] = (group_df.avg_backlog / bench_backlog) - 1
                group_df["rel_bench_overcap"] = (group_df.avg_overcap / bench_overcap) - 1
                
                savings_df_list.append(group_df)
        
        savings_df = pd.concat(savings_df_list, ignore_index=True)
    
    else:
        df["rel_actual_cost"] = (df.avg_cost / df[df.model == 'actual']['avg_cost'].values[0]) - 1
        df["rel_bench_cost"] = (df.avg_cost / df[df.model == 'L_4_sarimax']['avg_cost'].values[0]) - 1
        df["rel_actual_backlog"] = (df.avg_backlog / df[df.model == 'actual']['avg_backlog'].values[0]) - 1
        df["rel_bench_backlog"] = (df.avg_backlog / df[df.model == 'L_4_sarimax']['avg_backlog'].values[0]) - 1
        df["rel_bench_overcap"] = (df.avg_overcap / df[df.model == 'L_4_sarimax']['avg_overcap'].values[0]) - 1

        savings_df = df.copy()
    
    return savings_df


# Calc Savings

def calc_savings_weeks(df_w, scenario=None):
    
    # Create an empty list to store DataFrames
    df_list = []
    
    if scenario == None:
    
        # Loop through each unique week
        for week in df_w.week.unique():

            # Slice dataframe for the current week
            df = df_w[df_w["week"] == week].copy()

            # Apply calc_savings function and add results to df_list
            df_list.append(calc_savings(df))

        # Concatenate all dataframes in df_list
        savings_df_weekly = pd.concat(df_list)
    
    else:
        
        for week in df_w.week.unique():
            
            df = df_w[df_w["week"] == week].copy()
            df_list.append(calc_savings(df, group=True, scenario=scenario))
        
        # Concatenate all dataframes in df_list
        savings_df_weekly = pd.concat(df_list)
        
    return savings_df_weekly
```

```python
def plot_scenario_weeks(scenario_weeks_df, model_name=None, scenario=None):

    # Set Seaborn style
    sns.set(style="whitegrid")
    sns.set_context("paper")

    # Filter the DataFrame
    df = scenario_weeks_df[scenario_weeks_df["model"] == model_name].copy()

    # Create a categorical column with the week number as string
    df['week_str'] = 'Week ' + df['week'].astype(str)

    # Order weeks
    week_order = sorted(df['week_str'].unique())

    # Create the figure
    plt.figure(figsize=(10, 6))

    # Create the bar plot
    bar_plot = sns.barplot(x="week_str", y="rel_bench_cost", data=df, order=week_order, ci=None, hue=scenario)

    # Add labels and title
    plt.xlabel('Simulated Weeks (400 Scenarios per Week)', fontsize=12)
    plt.ylabel('Relative Average Bench Cost in %', fontsize=12)
    plt.title('Relative to Benchmark Cost for Best LightGBM Model per Simulated Week', fontsize=14)

    # Add mean line
    mean_rel_bench_cost = df['rel_bench_cost'].mean()
    plt.axhline(mean_rel_bench_cost, color='red', linestyle='--')
    plt.text(len(df.week_str.unique()), mean_rel_bench_cost, 'Mean: {:.2f}'.format(mean_rel_bench_cost), color='red')

    plt.tight_layout()
    plt.show()
```
