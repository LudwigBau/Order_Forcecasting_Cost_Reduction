```python
# Create Date Column
df["date"] = pd.to_datetime(df["order_datetime"]).dt.date
df["date"] = pd.to_datetime(df.date)
```

```python
# Get Basic Info
def var_basic_info(df):
    print(df.info())
    print(f"\nMax value {df.max()}")
    print(f"Min value: {df.min()}")

var_basic_info(df["date"])
```

```python
# Total Sales per Weekdays
weekdays = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
df_week_days = df.groupby(df["date"].dt.day_name()).mean().reindex(weekdays)
#df_week_days.reindex(weekdays).head()

# Total Sales per Week of year
df_week = df.groupby(df["date"].dt.week).mean()

# Total Sales per Month 
df_month = df.groupby(df["date"].dt.month).mean()

# Total Sales per Day of Year
df_day_year = df.groupby(df["date"].dt.dayofyear).mean()
```
